```python
import pandas as pd
from pathlib import Path

# Download the .xlsx from Google Drive and place it in this directory
file_path = Path("data") / "GCF NO.xlsx"

df = pd.read_excel(file_path, dtype={"zipcode": str})
df.columns = df.columns.str.strip().str.lower()
df["zipcode"] = df["zipcode"].str.strip().str.zfill(4)
df["weekdays"] = df["weekdays"].astype(str).str.strip()

# Validate that all zipcodes are exactly 4 characters
invalid = df[df["zipcode"].str.len() != 4]
if not invalid.empty:
    print(f"WARNING: {len(invalid)} zipcodes are not 4 characters:")
    print(invalid[["zipcode", "city"]].to_string(index=False))
else:
    print("All zipcodes are 4 characters.")

df.head(10)
```

```python
# Group rows by weekday combination
grouped = df.groupby("weekdays")

for combo, group in grouped:
    print(f"Weekdays: {combo} — {len(group)} rows")
    print(group[["zipcode", "city", "zone", "depot"]].to_string(index=False))
    print()
```

```python
# Print all unique weekday combinations
unique_combos = sorted(df["weekdays"].unique())

print(f"{len(unique_combos)} unique weekday combinations:\n")
for combo in unique_combos:
    print(combo)
```

```python
def zipcodes_for_weekdays(combo: str):
    """Print all zipcodes matching the given weekday combination (e.g. '2,7')."""
    match = df[df["weekdays"] == combo]
    if match.empty:
        print(f"No rows found for weekdays '{combo}'")
        return
    print(f"Zipcodes with weekdays '{combo}' ({len(match)} rows):\n")
    print(match[["zipcode", "city", "zone", "depot"]].to_string(index=False))


# Example usage:
zipcodes_for_weekdays("2,7")
```

```python
import json

output = {}
for combo, group in df.groupby("weekdays"):
    output[combo] = {"zipcodes": ",".join(sorted(group["zipcode"].tolist()))}

output_path = "output/delivery_zones.json"
with open(output_path, "w") as f:
    json.dump(output, f, indent=2, ensure_ascii=False)

print(f"Written to {output_path}")
print(json.dumps(output, indent=2, ensure_ascii=False))
```
